<h1>Introduction</h1>

This project does some statistical analysis on a star wars survey and compares results over different groups of people. 

```python
%matplotlib inline
from matplotlib import pyplot as plt
import pandas as pd
star_wars = pd.read_csv("star_wars.csv", encoding="ISO-8859-1")
star_wars.head()
```

```python
star_wars.columns 
```

```python
nulls=star_wars[pd.isnull(star_wars['RespondentID'])]
nulls
```

<h2>Null values</h2>

I found out that there is one null value in star_wars for respondentID that seems to be placeholder or have default values for each column. So, time to remove that.


```python
star_wars=star_wars[pd.notnull(star_wars['RespondentID'])]
star_wars.head(1)
```

<h1>Converting to boolean series</h1>

With a mapping I converted the first two columns to boolean for better and easier calculation later on.

```python
yes_no = {
    "Yes": True,
    "No": False
}

star_wars['Have you seen any of the 6 films in the Star Wars franchise?']=star_wars['Have you seen any of the 6 films in the Star Wars franchise?'].map(yes_no)
star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?']=star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'].map(yes_no)
star_wars.head()
```

<h2>More conversion</h2>

```python
import numpy as np

maps={
    "Star Wars: Episode I  The Phantom Menace": True,
    'Star Wars: Episode II  Attack of the Clones': True,
    'Star Wars: Episode III  Revenge of the Sith': True,
    'Star Wars: Episode IV  A New Hope': True,
    'Star Wars: Episode V The Empire Strikes Back': True,
    'Star Wars: Episode VI Return of the Jedi':True,
    np.nan: False
}
for col in star_wars.columns[3:9]:
    star_wars[col]=star_wars[col].map(maps)

star_wars = star_wars.rename(columns={
    "Which of the following Star Wars films have you seen? Please select all that apply.": "seen_1", 
    "Unnamed: 4": "seen_2", "Unnamed: 5": "seen_3", "Unnamed: 6": "seen_4", "Unnamed: 7": "seen_5", "Unnamed: 8": "seen_6"
    
})

star_wars.head()
```

```python
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)

star_wars = star_wars.rename(columns={
"Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.": "ranking_1",
    "Unnamed: 10": "ranking_2", "Unnamed: 11": "ranking_3", "Unnamed: 12": "ranking_4", "Unnamed: 13": "ranking_5", "Unnamed: 14": "ranking_6"
    
})
```

```python
cols=["ranking_1", "ranking_2", "ranking_3", "ranking_4", "ranking_5", "ranking_6"]

means=[]
for col in cols:
    mean=star_wars[col].mean()
    means.append(mean)
```

```python
import numpy as np
x=np.arange(6)
plt.bar(x, means, tick_label=cols, align='center') 
plt.xticks(rotation=90)
```

<h1>Results of user rankings</h1>

So far we have converted responses to true and false, renamed columns to make them more descriptive, and calculuated the means of each film. The data shows that The Empire Strikes Back is the favorite film of the series, and 4-6 all rate higher than 1-3. Revenge of the Sith is the least favorite movie. The respondents scores match critic and user reviews for the films, so it's not surprising to see these results.


```python
cols=["seen_1", "seen_2", 'seen_3', 'seen_4', 'seen_5', 'seen_6']

sums=[]
for col in cols:
    asum=star_wars[col].sum()
    sums.append(asum)
```

```python
plt.bar(x, sums, tick_label=cols, align='center') 
plt.xticks(rotation=90)
```

<h1>Results of how many people saw each movie</h1>

More people have seen 4-6 than 1-3, although few have seen 4 in comparison to 5 and 6.
The results correlate with the rankings rather well- the more times the movie had been watched, the better the ranking. So we don't know if Revenge of the Sith was the least favorite movie out of everyone who's seen it or only because it was the movie the least people saw.
However, we could find that out with the data we have.

```python
seen_all_6=pd.DataFrame()
for col in star_wars.columns[3:8]:
    seen_all_6=star_wars[star_wars[col]==True]
    
seen_all_6.shape
```

```python
cols=["ranking_1", "ranking_2", "ranking_3", "ranking_4", "ranking_5", "ranking_6"]

means=[]
means_seen_all_6=[]
for col in cols:
    mean=star_wars[col].mean()
    mean_all_6=seen_all_6[col].mean()
    means.append(mean)
    means_seen_all_6.append(mean_all_6)

x=np.arange(6)
fig, axarr=plt.subplots(2, sharex=True)
axarr[0].bar(x, means, tick_label=cols, align='center')
axarr[1].bar(x, means_seen_all_6, tick_label=cols, align='center')
plt.xticks(rotation=90)
axarr[0].set_title('Means for whole dataset')
axarr[1].set_title('Means for people who saw all 6 movies')
```

<h1>Comparison between people who saw all 6 and the whole dataset</h1> 

The results are surprisingly similar.

```python
males = star_wars[star_wars["Gender"] == "Male"]
females = star_wars[star_wars["Gender"] == "Female"]
```

<h1>Males Rankings</h1>

```python
cols=["ranking_1", "ranking_2", "ranking_3", "ranking_4", "ranking_5", "ranking_6"]

means=[]
for col in cols:
    mean=males[col].mean()
    means.append(mean)
    
x=np.arange(6)
plt.bar(x, means, tick_label=cols, align='center') 
plt.xticks(rotation=90)
```

<h2>Explanation</h2>

Remember that a higher ranking means a worse score for that movie. The lower the ranking, the better the movie.

<h1>Males Seen</h1>

```python
cols=["seen_1", "seen_2", 'seen_3', 'seen_4', 'seen_5', 'seen_6']

sums=[]
for col in cols:
    asum=males[col].sum()
    sums.append(asum)

plt.bar(x, sums, tick_label=cols, align='center') 
plt.xticks(rotation=90)

print(np.ptp(sums))
```

The range of Males seen is 75. Not a big amount when we compare to females. 

<h1>Females Rankings</h1>

```python
cols=["ranking_1", "ranking_2", "ranking_3", "ranking_4", "ranking_5", "ranking_6"]

means=[]
for col in cols:
    mean=females[col].mean()
    means.append(mean)
    
x=np.arange(6)
plt.bar(x, means, tick_label=cols, align='center') 
plt.xticks(rotation=90)
```

Already we see a big difference between male and female rankings. There's more extremes here.

<h1>Females Seen</h1>

```python
'FEMALES SEEN'
cols=["seen_1", "seen_2", 'seen_3', 'seen_4', 'seen_5', 'seen_6']

sums=[]
for col in cols:
    asum=females[col].sum()
    sums.append(asum)

plt.bar(x, sums, tick_label=cols, align='center') 
plt.xticks(rotation=90)

print(np.ptp(sums))
```

<h2>Range of females</h2>

The range for females seen is almost twice as large as males' range. Maybe this can also explain the greater variation in rankings for females. 

<h1>Analysis and Conclusion</h1> 

For males, we see that they liked 4-6 better as a group than 1-3 by a large margin. For females we don't see that pattern as sharply defined; females still liked The Empire Strikes Back the most but only liked 6 a bit more than 1-3 and 4 was about even with 1.
There a larger range in the number of females that have seen the Star Wars films than males: range is 131 for females and only 75 for males. Also relatively few females have seen A New Hope compared to the rest of the movies.
